```python
from tkinter import Tk, Label, Button, Entry, IntVar, END, W, E #Tkinter 위젯 불러오기

def update_add(): #더하기
  update("add") #더한 값 업데이트

def update_subtract(): #빼기
  update("subtract") #뺀 값 업데이트

def update_reset(): #초기화
  update("reset") #초기화한 값 업데이트

window = Tk() #모듈 만들기
total = 0 #초기화
sum = Label(window) 
sum.grid(row=0, column=1, columnspan=2) #행, 열, 폭

label = Label(window, text="현재 합계:") #라벨 표시될 공간과 내용 설정
label.grid(row=0, column=0) #위치 설정

entry = Entry(window) #한 줄의 텍스트를 입력 받는다
entry.grid(row=1, column=0, columnspan=3)

add_button = Button(window, text="더하기(+)", command=update_add) #더하기 버튼 설정
subtract_button = Button(window, text="빼기(-)", command=update_subtract) #빼기 버튼 설정
reset_button = Button(window, text="초기화", command=update_reset) #초기화 버튼 설정

add_button.grid(row=2, column=0) #더하기 버튼 위치 설정
subtract_button.grid(row=2, column=1) 
reset_button.grid(row=2, column=2) #초기화 버튼 위치 설정

def update(method): #업데이트 방법
  global total
  if method == "add": #더하는 방법
    total += int(entry.get()) # 공백 칸에서 얻어진 정수형 더해준다
  elif method == "subtract": # 빼는 법
    total -= int(entry.get()) #공백 칸에서 얻어진 정수형 빼준다
  else: # 그 외/ 초기화
    total = 0 #초기화
  sum['text'] = str(total) 
  entry.delete(0, END) #공백 칸 삭제 0부터 끝까지

window.mainloop() #윈도우 창을 윈도우를 종료할 때까지 유지
```

```python
import random #랜덤 모듈 불러오기
from tkinter import * #Tkinter 위젯 불러오기

window = Tk() #모듈 만들기
secret_number = random.randint(1, 100) #비밀 숫자 1부터 100까지 랜덤 지정
guess = None
num_guesses = 0

def guess_number(): #추측 값
  global num_guesses
  guess = int(entry.get())
  num_guesses += 1
  if guess == secret_number: # 답과 추측한 값이 같다면
    message = "축하합니다!!" #축하합니다
  elif guess < secret_number: # 답이 추측한 값보다 크다면
    message = "너무 낮아요!!" #너무 낮아요
  else: #답이 추측한 값보다 작으면
    message = "너무 높아요!!" #너무 높아요
  label['text']= message #메세지를 표시

def reset():
  global num_guesses
  entry.delete(0, END)
  secret_number = random.randint(1, 100)
  guess = 0
  num_guesses = 0
  message = "1부터 100사이의 숫자를 추측하시오"
  label['text']= message #메세지를 표시

message = "1부터 100사이의 숫자를 추측하시오"
label = Label(window, text=message) #라벨이 표시될 공간과 내용 설정
entry = Entry(window) #한 줄의 텍스트를 입력 받는 

guess_button = Button(window, text="숫자를 입력", command=guess_number) #guess button 설정
reset_button = Button(window, text="게임을 다시 실행", command=reset) #reset 버튼 설정

label.grid(row=0, column=0, columnspan=2, sticky=W+E) #라벨 격자
entry.grid(row=1, column=0, columnspan=2, sticky=W+E) #입력 받는 칸 격자
guess_button.grid(row=2, column=0) #guess 버튼 격자
reset_button.grid(row=2, column=1) #reset 버튼 격자

window.mainloop() #윈도우 창을 윈도우를 종료할 때까지 유지
```

```python
from tkinter import * #tkinter 위젯 불러오기

def do_convert(): #전환
  inch_val = float(cvt_from.get()) #float: 숫자나 문자열을 실수로 변환/ 전환할 값 받기
  centimeters_val = inch_val * 2.54  #인치를 센티미터로 변환하는 식
  cvt_to.set('%s 센티미터' % centimeters_val) #문자열

root = Tk() #모듈 만들기
cvt_from = StringVar() #StringVar()로 정의된 변수는 set 매서드를 이용해줘야 한다
cvt_to = StringVar()

lbl = Label(root, text='인치를 센티미터로 변환하는 프로그램:') #라벨 표시될 공간과 내용 설정
lbl.grid(row=0, column=0, columnspan=2) #위의 내용 담을 공간 (격자)설정
from_lbl = Label(root, text='인치를 입력하시오:') #초기값 라벨 설정
from_lbl.grid(row=1, column=0) #초기값 격자 설정
from_entry = Entry(root, textvariable=cvt_from) #초기값 공백칸 라벨 설정
from_entry.grid(row=1, column=1) #초기 값 격자 설정

to_lbl = Label(root, text='변환결과:') #전환값 라벨 설정
to_lbl.grid(row=2, column=0) #전환값 격자 설정
result_lbl = Label(root,textvariable=cvt_to) #변환값 결과 라벨 표시 공간, 내용 설정
result_lbl.grid(row=2, column=1) #변환값 결과 격자 설정

convert_btn = Button(root,text='변환!', command=do_convert) #전환 버튼 설정
convert_btn.grid(row=3, column=1) #전환 버튼 격자 설정
root.mainloop() #윈도우 창을 윈도우를 종료할 때까지 유지
```

```python
from tkinter import * #tkinter 위젯 불러오기
fields = '이름', '직업', '국적' 

def fetch(entries): 
  for entry in entries:
    field = entry[0]
    text = entry[1].get()
    print('%s: "%s"' % (field, text)) 

def makeform(root, fields): 
  entries = []
  for field in fields:
    row = Frame(root)
    lab = Label(row, width=15, text=field)
    ent = Entry(row)
    row.pack(side=TOP, fill=X)
    lab.pack(side=LEFT)
    ent.pack(side=RIGHT, expand=YES, fill=X)
    entries.append((field, ent))
  return entries

root = Tk()
ents = makeform(root, fields)
root.bind('<Return>', (lambda event, e=ents: fetch(e))) 
b1 = Button(root, text='보여주기',command=(lambda e=ents: fetch(e)))
b1.pack(side=LEFT, padx=5, pady=5)
b2 = Button(root, text='종료하기', command=root.quit)
b2.pack(side=LEFT, padx=5, pady=5)
root.mainloop()
```

```python
from random import * #랜덤 모듈 불러오기
from tkinter import * #tkinter 위젯 불러오기

# 선택 하는 부분
def user_choice_rock():
  user_choice = "rock"
  turn(user_choice)
  user_image.configure(image=rock_image)
def user_choice_paper():
  user_choice = "paper"
  turn(user_choice)
  user_image.configure(image=paper_image)
def user_choice_scissors():
  user_choice = "scissors"
  turn(user_choice)
  user_image.configure(image=scissors_image)

# 게임부분
def turn(user_choice):
  oppo = ['rock', 'paper', 'scissors']
  oppo_choice=oppo[randint(0,2)]
  if(oppo_choice=='rock'):
    oppo_image.configure(image=rock_image)
    if(user_choice=='paper'):
      turn_result.configure(text="사용자 승!", fg="green")
      compare.configure(text=">>>>>")
    elif(user_choice=='scissors'):
      turn_result.configure(text="컴퓨터 승!", fg="red")
      compare.configure(text="<<<<<")
    else:
      turn_result.configure(text="무승부", fg="gray")
      compare.configure(text="=====")
 
  elif(oppo_choice=='paper'):
    oppo_image.configure(image=paper_image)
    if(user_choice=='scissors'):
      turn_result.configure(text="사용자 승!", fg="green")
      compare.configure(text=">>>>>")
    elif(user_choice=='rock'):
      turn_result.configure(text="컴퓨터 승!", fg="red")
      compare.configure(text="<<<<<")
    else:
      turn_result.configure(text="무승부", fg="gray")
      compare.configure(text="=====")
 
  elif(oppo_choice=='scissors'):
    oppo_image.configure(image=scissors_image)
    if(user_choice=='rock'):
      turn_result.configure(text="사용자 승!", fg="green")
      compare.configure(text=">>>>>")
    elif(user_choice=='paper'):
      turn_result.configure(text="컴퓨터 승!", fg="red")
      compare.configure(text="<<<<<")
    else:
      turn_result.configure(text="무승부", fg="gray")
      compare.configure(text="=====")
 
# 메인 프로그램
main_window = Tk()
rock_button = Button(main_window, width=20, text="바위", justify=CENTER, 
command=user_choice_rock, activebackground='black', activeforeground='white')
paper_button = Button(main_window, width=20, text="보", justify=CENTER, 
command=user_choice_paper, activebackground='black', activeforeground='white')
scissors_button = Button(main_window, width=20, text="가위", justify=CENTER, 
command=user_choice_scissors, activebackground='black', activeforeground='white')
rock_image = PhotoImage(file="d:/rock.gif")
paper_image = PhotoImage(file="d:/paper.gif")
scissors_image = PhotoImage(file="d:/scissors.gif")
user_image = Label(text="사용자", image=rock_image)
user_image.image = rock_image
compare = Label(main_window, justify=CENTER, font=("Helvetica", 30))
oppo_image = Label(text="컴퓨터",image=paper_image)
oppo_image.image = paper_image
turn_result = Label(main_window, width=20, justify=CENTER, font=("Helvetica", 20))

# 그리드 생성
rock_button.grid(row=5, column=1)
paper_button.grid(row=5, column=2)
scissors_button.grid(row=5, column=3)
user_image.grid(row=3, column=1)
compare.grid(row=3, column=2)
oppo_image.grid(row=3, column=3)
turn_result.grid(row=4, column=2)

# GUI화면 루프처리
main_window.mainloop()
```
